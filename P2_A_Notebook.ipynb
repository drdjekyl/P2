### 2.2 Recherche du premier jeu d'indicateurs pertinents pour les années 2010 à 2015 classé par pays

```python
countries_nb = i = 160

past = groupby_Years_Indicators.filter(lambda x: x['2010'].count() > i
                                and x['2011'].count() > i
                                and x['2012'].count() > i
                                and x['2013'].count() > i
                                and x['2014'].count() > i
                                and x['2015'].count() > i
                                      )

past['Indicator Name'].unique() # Affiche la liste des indicateurs disponibles

#taux de NaN/pays intéressant et indicateurs pertinents,
#en regard des choix stratégiques que cherche à réaliser l'entreprise Academy
```

```python
past.drop(columns=['2016', '2017', '2020', '2025', '2030'])
#Nettoyage pour obtenir les années souhaitées : a savoir 2020 à 2030
```

```python
indic_a = 'Population, ages 15-24, total'
indic_b = 'Population of the official age for upper secondary education, both sexes (number)'
indic_c = 'Official entrance age to upper secondary education (years)'
indic_d = 'GDP per capita (current US$)'

# Sélection des indicateurs
```

```python
df_indic_a = past[past['Indicator Name'] == indic_a]
df_indic_b = past[past['Indicator Name'] == indic_b]
df_indic_c = past[past['Indicator Name'] == indic_c]
df_indic_d = past[past['Indicator Name'] == indic_d]

past_population = pd.concat([df_indic_a, df_indic_b, df_indic_c, df_indic_d])
#Création d'un nouveau dataframe
past_population = past_population.pivot_table(index=['Country Name'], columns=['Indicator Name'])

#Retenons ces données comme indicateurs de l'évolution de la population par pays susceptible de vouloir souscrire
#aux offres de formation d'Academy !!
```

```python
top5 = past_population['2015'].sort_values(by=[indic_a], ascending = False)

droplist = ['World','Upper middle income','High income','Low income','Middle income','Lower middle income']

for i in droplist:
    top5.drop(index=i, inplace=True)
```

```python
top5_final = top5.head(5)
top5_final
```

#### Nombre de clients potentiels classé par pays en 2015

```python
fig,ax1 = plt.subplots(figsize=(10,8))

y1 = top5_final['Population, ages 15-24, total'].sort_values(ascending=True)
r1 = range(len(y1))
xticks = y1.index

fig.suptitle('TOP 5 PAYS : Population (15-24 ans) en 2015', fontsize=16)
plt.xticks(range(len(y1)), xticks)
ax1.bar(r1, y1);

plt.savefig("img3")
```

```python
top5_Upp = past_population['2015'].sort_values(by=[indic_b], ascending = False)

for i in droplist:
    top5_Upp.drop(index=i, inplace=True)
top5_Upp = top5_Upp.head(5)
```

```python
fig,ax1 = plt.subplots(figsize=(10,8))

y1 = top5_Upp['Population of the official age for upper secondary education, both sexes (number)'].sort_values(
    ascending=True)
r1 = range(len(y1))
xticks = y1.index

fig.suptitle('TOP 5 PAYS : Population ayant l\'âge officiel d\'entrer au Upper Secondary en 2015', fontsize=16)
plt.xticks(range(len(y1)), xticks)
ax1.bar(r1, y1);

plt.savefig("img4")
```

```python
GDP_scale = [1000, 20000, 40000]
data_donut = []

for i in GDP_scale:
    data_donut.append(top5[top5['GDP per capita (current US$)'] > i].head(5))

col = ['GDP per capita (current US$)', 'Official entrance age to upper secondary education (years)', 
       'Population, ages 15-24, total']

for i in range(3):
    for j in col:
        data_donut[i].drop(columns=j, inplace=True)
```

```python
theme1 = ['#BAF73E','#47F73E','#A63EF7','#3EAAF7','#3E51F7']
theme2 = ['#A63EF7','#FD465A','#F8B531','#F45FA5', '#C572F3']
theme3 = ['#A63EF7','#F8B531','#BDF57A','#F5321F','#F7F614']
fig,(ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(15,5), constrained_layout=True)

ax1.set_title('PIB par habitant > 1000$', size=16)
ax2.set_title('PIB par habitant > 20000$', size=16)
ax3.set_title('PIB par habitant > 40000$', size=16)

def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "nb : {:d}".format(absolute)

wedges1, texts1, autotexts1 = ax1.pie(data_donut[0], autopct=lambda pct: func(pct, data_donut[0]),
                                   textprops=dict(color='k'), radius=1.2, colors=theme1)
ax1.legend(wedges1, data_donut[0].index, title="Pays", loc="best")

wedges2, texts2, autotexts2 = ax2.pie(data_donut[1], autopct=lambda pct: func(pct, data_donut[1]),
                                   textprops=dict(color='k'), radius=1.2, colors=theme2)
ax2.legend(wedges2, data_donut[1].index, title="Pays", loc="best")

wedges3, texts3, autotexts3 = ax3.pie(data_donut[2], autopct=lambda pct: func(pct, data_donut[2]), 
                                   textprops=dict(color='b'), radius=1.2, colors=theme3)
ax3.legend(wedges3, data_donut[2].index, title="Pays", loc="best")

fig.suptitle('TOP 5 PAYS : Nombre d\'étudiants en âge d\'aller au "Upper Secondary" en 2015', size=16, weight='bold')
plt.setp(autotexts1, size=11, color='k')
plt.setp(autotexts2, size=11, color='k')
plt.setp(autotexts3, size=11, color='k')

plt.show();

plt.savefig("img")
```

# 3. Une autre approche pour les indicateurs statistiques généraux

```python
col = [0,1,2,3,49]
data_stats = data.iloc[:, col]
data_stats.info()

pd.pivot_table(data_stats, index=['Indicator Code', 'Indicator Name']).dropna(axis='index')
#Sélection de l'année 2015 car année avec suffisamment d'indicateurs pour illustrer quelques ordres de grandeur
```

```python
data_stats = pd.pivot_table(data_stats, index=['Indicator Code', 'Indicator Name', 'Country Name'])
```

```python
data_stats = data_stats.stack()
#possibilité d'avoir une approche par indicateur, par pays et par année
#Nécessite de retrouver les indicateurs par code afin de trouver les clients potentiels
```

```python
a = data_stats['NY.GDP.PCAP.CD'].unstack().sort_values(by=['2015'], ascending=False)
# (PIB par hab)
b = data_stats['IT.NET.USER.P2'].unstack().sort_values(by=['2015'], ascending=False)
# (internet pour 100 hab)
c = data_stats['SE.ADT.1524.LT.ZS'].unstack().sort_values(by=['2015'], ascending=False)
# (15-24 ans lettrés)
d = data_stats['SP.POP.TOTL'].unstack().sort_values(by=['2015'], ascending=False)
# (pop totale)
e = data_stats['SP.POP.1524.TO.UN'].unstack().sort_values(by=['2015'], ascending=False)
# (population 15-24)

data_stats = pd.concat([a, b, c, d, e]).pivot_table(index=['Country Name'], columns=['Indicator Name'])
```

```python
var = data_stats.sort_values([('2015','Population, total')], ascending=False)

continents = var.iloc[0:24,:]
continents = continents.drop(index='China')
continents = continents.drop(index='India')
continents.drop(columns=[('2015','Population, ages 15-24, total')], inplace=True)
continents.drop(columns=[('2015','Youth literacy rate, population 15-24 years, both sexes (%)')], inplace=True)
```

```python
fig,ax1 = plt.subplots(figsize=(10,8))

y1 = continents['2015','GDP per capita (current US$)'].sort_values(ascending=True)
r1 = range(len(y1))
yticks = y1.index

fig.suptitle('Répartition du PIB/hab par continent en 2015, en US$', fontsize=16)
plt.yticks(range(len(y1)), yticks)
ax1.barh(r1, y1);

plt.savefig("img7")
```

```python
median = continents['2015','GDP per capita (current US$)'].median()
mean = continents['2015','GDP per capita (current US$)'].mean()
std = continents['2015','GDP per capita (current US$)'].std()

print('Médiane du PIB par habitants en 2015 : {:.2f}'.format(median))
print('Moyenne du PIB par habitants en 2015 : {:.2f}'.format(mean))
print('Ecart type du PIB par habitants en 2015 : {:.2f}'.format(std))
```

```python
for i in continents.index:
    var.drop(index=i, inplace=True)
```

```python
countries = var
i = countries['2015', 'GDP per capita (current US$)'].dropna()
j = countries['2015', 'Internet users (per 100 people)'].dropna()
```

```python
green_diamond = dict(markerfacecolor='g', marker='D')
fig1, ax1 = plt.subplots(figsize=(20,3))
ax1.set_title('Répartition du PIB par habitant (US $), par pays en 2015')
plt.boxplot(i, vert=False, showfliers=False, notch=True);

plt.savefig("img8")
```
